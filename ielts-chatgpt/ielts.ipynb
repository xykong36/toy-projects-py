```python
"""
TODO:
call API 的局限性

save the result to h

"""
from youtube_transcript_api import YouTubeTranscriptApi
import re
import openai
import os



# summarize one video

def parse_video_url(video_url):
    video_id = re.search(r'(?<=v=)[\w-]+', video_url).group()
    return video_id

# get the transcript for one video
def get_transcript_for_video(video_id):
    res = YouTubeTranscriptApi.get_transcript(video_id)
    text_list = [item['text'] for item in res]
    transcript_str = "".join(text_list)
    return transcript_str

# call open ai to summarize
def summarize_video_transcript(transcript):
    summarize_prompt = "summarize the following text with 5 bullet points: "
    request_prompt = summarize_prompt + "" + transcript
    
    openai.api_key = "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"
    response = openai.Completion.create(
        engine="text-davinci-002",
        prompt=f"Q: {request_prompt}\nA:",
        max_tokens=3056,
        n=1,
        stop=None,
        temperature=0.4,
    )
    resps.append(response)
    # Extract the answer from the OpenAI API response
    answer = response.choices[0].text.strip()

    # Print the question and answer
    print(f"Q: {question}")
    print(f"A: {answer}")
```
